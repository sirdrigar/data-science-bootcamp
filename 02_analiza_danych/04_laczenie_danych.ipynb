##Import bibliotek

```python
import pandas as pd
pd.__version__
```

##Wczytywanie danych

```python
def fetch_finanacial_data(company='AMZN'):
  """
    This is function fetch stock market quotations.
  """
  import pandas_datareader.data as web
  return web.DataReader(name=company, data_source='stooq')

apple = fetch_finanacial_data('AAPL')
amazon = fetch_finanacial_data('AMZN')
google = fetch_finanacial_data('GOOGL')
uber = fetch_finanacial_data('UBER')
```

```python
apple.info()
```

```python
uber.info()
```

```python
apple.head()
```

```python
apple.columns = [ 'apple_' + col.lower() for col in apple.columns]
amazon.columns = [ 'amazon_' + col.lower() for col in amazon.columns]
google.columns = [ 'google_' + col.lower() for col in google.columns]
uber.columns = [ 'uber_' + col.lower() for col in uber.columns]
```

```python
apple.head()
```

##Konkatenacja danych

```python
df = pd.concat(objs=[apple, amazon, google, uber], axis=1)
df
```

```python
pd.set_option('display.float_format', lambda x: f'{x:.2f}')
df.describe().T
```

```python
df.corr()
```

```python
df.columns
```

```python
closes = [col for col in df.columns if col.endswith('_close') ]
closes
```

```python
df[closes].corr()
```

```python
closes = [col for col in df.columns if col.endswith('_close') or col.endswith('_volume')]
closes
```

```python
df[closes].corr()
```
